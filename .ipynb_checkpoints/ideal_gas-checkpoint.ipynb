```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sm

from matplotlib.animation import FuncAnimation
from IPython.display import HTML 
```

```python
# Constants
box_size       = 1.0
num_particles  = 40
particle_speed = 0.05
time_steps     = 200

# Initial positions and velocities of particles
positions  = np.random.rand(num_particles, 2) * box_size
velocities = np.random.rand(num_particles, 2) * particle_speed

plt.scatter(positions[:, 0], positions[:, 1])
plt.xlabel('x')
plt.ylabel('y')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Plot setup
fig, ax = plt.subplots()
scatter = ax.scatter(positions[:, 0], positions[:, 1], s=20)
ax.set_xlim(0, box_size)
ax.set_ylim(0, box_size)

# Update function
def update(frame):
    global positions, velocities

    positions += velocities

    # Bounce logic
    for i in range(num_particles):
        for j in range(2):
            if positions[i, j] < 0 or positions[i, j] > box_size:
                velocities[i, j] *= -1
                positions[i, j] = np.clip(positions[i, j], 0, box_size)

    scatter.set_offsets(positions)
    return scatter,

# Create animation
ani = FuncAnimation(fig, update, frames=time_steps, interval=25, blit=True)

# Display in Jupyter Notebook
HTML(ani.to_html5_video())
```
